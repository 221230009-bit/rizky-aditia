```python

# ðŸ›’ LATIHAN 3: ANALISIS DATASET RETAIL (Pandas)

import numpy as np
import pandas as pd

'''TODO: Analisis Dataset Retail'''
# Create sample retail dataset
np.random.seed(42)
dates = pd.date_range('2023-01-01', '2023-03-31', freq='D')

retail_data = pd.DataFrame({
    'date': np.random.choice(dates, 1000),
    'product_id': np.random.randint(1, 11, 1000),
    'quantity': np.random.randint(1, 10, 1000),
    'price': np.random.uniform(10, 100, 1000),
    'customer_id': np.random.randint(1, 101, 1000)
})
retail_data['revenue'] = retail_data['quantity'] * retail_data['price']

# 1ï¸âƒ£ Hitung total revenue per product
revenue_per_product = retail_data.groupby('product_id')['revenue'].sum().reset_index().sort_values(by='revenue', ascending=False)

# 2ï¸âƒ£ Temukan top 5 customers berdasarkan total spending
top_customers = (
    retail_data.groupby('customer_id')['revenue']
    .sum()
    .reset_index()
    .sort_values(by='revenue', ascending=False)
    .head(5)
)

# 3ï¸âƒ£ Analisis time series - revenue per hari
daily_revenue = (
    retail_data.groupby('date')['revenue']
    .sum()
    .reset_index()
    .sort_values(by='date')
)

# 4ï¸âƒ£ Buat fungsi untuk detect anomalies dalam quantity
def detect_quantity_anomalies(df, threshold=2):
    mean_qty = df['quantity'].mean()
    std_qty = df['quantity'].std()
    df['z_score'] = (df['quantity'] - mean_qty) / std_qty
    anomalies = df[np.abs(df['z_score']) > threshold]
    return anomalies

anomalies = detect_quantity_anomalies(retail_data)

# âœ… Validasi
assert len(revenue_per_product) <= 10, "Should have max 10 products"
assert len(top_customers) == 5, "Should have top 5 customers"

print("âœ… Pandas operations completed successfully!\n")

# ðŸ“Š Tampilkan hasil ringkas
print("=== TOTAL REVENUE PER PRODUCT ===")
print(revenue_per_product.head(), "\n")

print("=== TOP 5 CUSTOMERS (Total Spending) ===")
print(top_customers, "\n")

print("=== DAILY REVENUE (Preview) ===")
print(daily_revenue.head(), "\n")

print("=== DETECTED ANOMALIES IN QUANTITY ===")
print(anomalies.head(), "\n")
```
